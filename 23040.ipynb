```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/KC Survey_All Years.csv')
df
```

```python
FactChecker = {"legal_addgps":2,"legal_gps1819":1,"legal_gps2122": 2,"legal_handheld":2,"legal_handheldstop":2,"legal_handheldspkr":2,"legal_911":1,"legal_camera":2, "legal_music1819":1, "legal_music2122":2, "legal_readstop":2, "legal_typestop":2, "legal_handfree":1, "legal_app1819":2, "legal_app2122":2, "legal_video":2}
FactChecker
#These are the TRUE values, what the law actually says
```

```python
def GetPerception(ColumnName):
#This will return a pandas dataframe that maps a certain column name that you want to compare to the perception of the individuals
#knowledge of driving laws. For example, if you wanted to compare education level to their perception of driving laws, you would call
#"GetPerception("Education")" and be given a dataframe with two columns, highest level of education and their percetage of correct perception
#Of Washington driving laws.



  finallist = [] #Initializes list where I will store dictionaries before I turn into Dataframe
  for index,row in df.iterrows():
  #I first initialize two dictionaries and a list that will be reset every row of the dataframe and appended to finallist above
    DictOfAverages = {} #Initializes Dictionary that will contain the ColumnName : ColumnNameValue and Perception : Average perception score
    ColumnNameValue = row[ColumnName]
    Perception = {}
    ListToAverage = []
    for item in FactChecker:
      Perception[item] = row[item]
      if Perception[item] == 3 or pd.isna(Perception[item]): #This removes values where the participant answered "Not sure" or didn't answer the question at all
        del Perception[item]
  #This for loop checks if their perception about a law is true or not and if it is, it counts as a 1, if not it counts as a zero
    for item in Perception:
      if Perception[item] == FactChecker[item]:
        ListToAverage.append(1)
      else:
        ListToAverage.append(0)
    if len(ListToAverage) != 0: #makes sure that there is at least one value in the list
      if pd.isna(ColumnNameValue) == False: #makes sure that they answered the question of education level
        Average = sum(ListToAverage)/len(ListToAverage)
        DictOfAverages[ColumnName] = ColumnNameValue
        DictOfAverages["Perception"] = Average
        DictOfAverages["year"] = row["year"]
    finallist.append(DictOfAverages)
  FinalDataFrame = pd.DataFrame(finallist).dropna() #converts the list of Dictionaries into a pandas dataframe, dropping the NaN columns (if the participant didn't answer either of the questions)
  return FinalDataFrame

GetPerception("Education")
  
```

```python
def GetYearMeans(dataframe, ColumnName):
#This function takes a dataframe and a ColumnName (same argument as GetPerception) and returns a pandas dataframe that combines the means
#of the perception per categorical variable and has a year column. Therefore it will be extremely easy to graph using plotly express
  y2018 = pd.DataFrame(dataframe[dataframe["year"] == 2018.0].groupby(ColumnName)["Perception"].mean()).reset_index()
  y2018['year'] = "2018"
  y2019 = pd.DataFrame(dataframe[dataframe["year"] == 2019.0].groupby(ColumnName)["Perception"].mean()).reset_index()
  y2019['year'] = "2019"
  y2021 = pd.DataFrame(dataframe[dataframe["year"] == 2021.0].groupby(ColumnName)["Perception"].mean()).reset_index()
  y2021['year'] = "2021"
  y2022 = pd.DataFrame(dataframe[dataframe["year"] == 2022.0].groupby(ColumnName)["Perception"].mean()).reset_index()
  y2022['year'] = "2022"
  combined = pd.concat([y2018, y2019, y2021, y2022])
  return combined
education = GetPerception("Education")
education = GetYearMeans(education, "Education")
education
```

```python
import plotly.express as px
import plotly.graph_objects as go
education = GetPerception("Education")
combined = GetYearMeans(education, "Education")
combined.replace(1, "High School", inplace=True)
combined.replace(2, "Some College", inplace=True)
combined.replace(3, "Associate's/Bachelor's", inplace=True)
combined.replace(4, "Master's Or Higher", inplace=True)
fig = px.bar(combined, x="Education", y="Perception", color="year", barmode='group')
fig.show()
```

```python
income = GetPerception("Income")
combined = GetYearMeans(income, "Income")
combined.replace(1, "<$40K", inplace=True)
combined.replace(2, "$40-80K", inplace=True)
combined.replace(3, "$80-120K", inplace=True)
combined.replace(4, "$120-160K", inplace=True)
combined.replace(5, ">$160K", inplace=True)
fig = px.bar(combined, x="Income", y="Perception", color="year", barmode='group')
fig.show()
```

```python
race = GetPerception("Race")
combined = GetYearMeans(race, "Race")
combined.replace(1, "Hispanic", inplace=True)
combined.replace(2, "Black", inplace=True)
combined.replace(3, "White", inplace=True)
combined.replace(4, "Asian", inplace=True)
combined.replace(5, "American Indian", inplace=True)
combined.replace(6, "Native Hawaiian", inplace=True)
combined.replace(7, "Multiracial", inplace=True)
combined.replace(8, "Other", inplace=True)
fig = px.bar(combined, x="Race", y="Perception", color="year", barmode='group')
fig.show()
```
